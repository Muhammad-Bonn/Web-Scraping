```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
response = requests.get("https://en.wikipedia.org/wiki/List_of_programming_languages")
```

```python
bs = BeautifulSoup(response.text, 'lxml')
data = bs.find_all("div", {"class": "div-col"})
```

```python
listed = str(data).split('</a></li>')
listed.pop(0)
```

```python
listed;
```

```python
for item in listed:
    if '22em' in str(item):
        listed.pop(listed.index(item))

links = []
titles = []

for i in range(len(listed)-1):
    list_of_title_and_link = listed[i].split('href=')[1].split('title')
    link = list_of_title_and_link[0].split('"')[1]
    links.append(link)
    title = list_of_title_and_link[1].split('"')[1]
    titles.append(title)
```

```python
df = pd.DataFrame({'Language': titles, 'Link': links})
for title in df["Language"]:
    if '(programming language)' in str(title):
        titleNew = title.strip(' (programming language)')
        df.loc[df['Language'] == title, 'Language'] = titleNew
df
```

```python
definitions_list = []
for link in df['Link']:
    link = "https://en.wikipedia.org"+ link
    res = requests.get(link)
    soup= BeautifulSoup(res.text, 'lxml')

    definition = soup.find('p').getText()
    definitions_list.append(definition)

df['Definition'] = definitions_list
df
```

```python
# save df to csv
df.to_csv('programming_languages_def.csv')
```
